# Visualización de Datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Diagramas de Dispersion (scatterplot)

# scatter graficos para correlación de variables

x = np.arange(2,20)
y1 = x**2
y2 = x**3

plt.scatter(x, y1, color='green', marker='*', label='x^2')
plt.scatter(x, y2, color='blue', marker='o', label='x^3')
plt.plot(x, y2, color='pink', label='line')
plt.title('Funciones Polinomicas', fontsize =14, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()

plt.text(5, 4000, 'texto', color='blue')
```

```python
# Histogramas

mean = 5
std = 3
n = 10000

x = np.random.normal(mean, std, n)

plt.hist(x, bins=20);
plt.title('Histograma de una variable aleatoria', fontsize =14, color='red')
plt.xlabel('x')
plt.ylabel('frecuencia')
plt.grid()
plt.legend()

plt.text(5, 1400, f'Mean: texto', color='blue')
```

```python
plt.boxplot(x, vert=False);
```

```python
# Subplot

fig, ax = plt.subplots(ncols=1, nrows=2, figsize=(5,10), sharex=True)

ax[0].hist(x);

ax[1].boxplot(x, vert=False);
```

```python
# Barras

classes = ['gatos', 'perros','ratones']
totales = [20, 35, 12]

plt.bar(classes, totales,);

```

```python
import seaborn as sns

data = pd.DataFrame({

    'Animales': classes,
    'Totales': totales
})

data.head()

sns.barplot(x='Animales', y='Totales', data=data)
```
