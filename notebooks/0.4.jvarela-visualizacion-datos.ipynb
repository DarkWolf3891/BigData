# Visualizaci√≥n de Datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Diagramas de Dispersion (scatterplot)

# scatter 

x = np.arange(2,20)
y1 = x**2
y2 = x**3

plt.scatter(x, y1, color='green', marker='*', label='x^2')
plt.scatter(x, y2, color='blue', marker='o', label='x^3')
plt.plot(x, y2, color='pink', label='line')
plt.title('Funciones Polinomicas', fontsize =14, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()

plt.text(5, 4000, 'texto', color='blue')
```
